# Using Box-Cox

```python
num_bc=num.copy()
```

```python
def boxcox_transform(df):
    numeric_cols = df.select_dtypes(np.number).columns
    _ci = {column: None for column in numeric_cols}
    for column in numeric_cols:
        df[column] = np.where(df[column]<=0, np.NAN, df[column]) 
        df[column] = df[column].fillna(df[column].mean())
        transformed_data, ci = stats.boxcox(df[column])
        df[column] = transformed_data
        _ci[column] = [ci] 
    return df, _ci
```

```python
num_bc, _ci = boxcox_transform(num_bc)
```

```python
for column in num_bc.columns:
    sns.distplot(num_bc[column], kde=True, color = 'royalblue')
    plt.show()
```

# Using Log-Transform

# Using Power Transform

```python
def power_transform(x, p):
    return x**p
```

```python
num_pt=num.copy()
```

```python
num_pt = num_pt.apply(lambda x: power_transform(x, p=2), axis=1)
```

```python
for column in num_pt.columns:
    sns.distplot(num_pt[column], kde=True, color = 'mediumorchid')
    plt.show()
```
